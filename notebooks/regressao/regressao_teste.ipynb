## Linear Regression

Neste bloco é treinado um modelo de regressão linear simples usando os dados de treino. Depois o modelo é validado com os dados de validação e por fim são somados os datasets de treinamento e validação, para criar um modelo final que será testado com os dados de teste.
Em seguida são calculadas métricas de desempenho para avaliar a qualidade do ajuste.

```python
# Model Training
lr_model = LinearRegression()
lr_model.fit( X_train, y_train )
```

```python
#Previsão sobre os dados de treinamento
y_pred_train = lr_model.predict( X_train )
# # R²
r2_squared_train = mt.r2_score(y_train, y_pred_train)
print("R²:", r2_squared_train)
# MSE 
mse_train = mt.mean_squared_error(y_train, y_pred_train)
print("MSE", mse_train)
# RMSE
rmse_train = np.sqrt(mse_train)
print("RMSE", rmse_train)
# MAE
mae_train = mt.mean_absolute_error(y_train, y_pred_train)
print("MAE", mae_train)
# MAPE
y_train_flat = y_train
y_pred_train_flat = y_pred_train.ravel()

mask = y_train_flat != 0
mape_train = np.mean(
    np.abs((y_train_flat[mask] - y_pred_train_flat[mask]) / y_train_flat[mask])
)
print("MAPE", mape_train)
```

```python
#Previsão sobre os dados de validação
y_pred_val = lr_model.predict( X_val )
# # R²
r2_squared_val = mt.r2_score(y_val, y_pred_val)
print("R²:", r2_squared_val)
# MSE 
mse_val = mt.mean_squared_error(y_val, y_pred_val)
print("MSE", mse_val)
# RMSE
rmse_val = np.sqrt(mse_val)
print("RMSE", rmse_val)
# MAE
mae_val = mt.mean_absolute_error(y_val, y_pred_val)
print("MAE", mae_val)
# MAPE
y_val_flat = y_val
y_pred_val_flat = y_pred_val.ravel()

mask = y_val_flat != 0
mape_val = np.mean(
    np.abs((y_val_flat[mask] - y_pred_val_flat[mask]) / y_val_flat[mask])
)
print("MAPE", mape_val)
```

```python
#Agora vamos treinar o modelo com os dados de treinamento e validação juntos para depois testar com os dados de teste
#  Model Training
model_last_lr = LinearRegression()
model_last_lr.fit( pd.concat( (X_train, X_val) ), np.concatenate((y_train, y_val)) )
```

```python
# Modelo validado com os dados de teste e treinada com os dados de treinamento e validação juntos
#Previsão sobre os dados de teste
y_pred_test = model_last_lr.predict( X_test )
# # R²
r2_squared_test = mt.r2_score(y_test, y_pred_test)
print("R²:", r2_squared_test)
# MSE 
mse_test = mt.mean_squared_error(y_test, y_pred_test)
print("MSE", mse_test)
# RMSE
rmse_test = np.sqrt(mse_test)
print("RMSE", rmse_test)
# MAE
mae_test = mt.mean_absolute_error(y_test, y_pred_test)
print("MAE", mae_test)
# MAPE
y_test_flat_dt = y_test
y_pred_test_dt = y_pred_test.ravel()

mask = y_test_flat_dt != 0
mape_test = np.mean(
    np.abs((y_test_flat_dt[mask] - y_pred_test_dt[mask]) / y_test_flat_dt[mask])
)
print("MAPE", mape_test)
```

## Decision Tree Regressor

Aqui é treinado um modelo de árvore de decisão para regressão.
Essa abordagem captura não linearidades e pode ser ajustada com `max_depth`.
Neste bloco é treinado um modelo de Decision Tree Regressor usando os dados de treino. Depois o modelo é validado com os dados de validação e por fim são somados os datasets de treinamento e validação, para criar um modelo final que será testado com os dados de teste.
Em seguida são calculadas métricas de desempenho para avaliar a qualidade do ajuste.

```python
#modelo Decision Tree Regressor inicial treinado com os dados de treinamento e parâmetros default
model_train_dt = tr.DecisionTreeRegressor()
model_train_dt.fit(X_train, y_train)
```

```python
#Previsão sobre os dados de treinamento
y_pred_train_dt = model_train_dt.predict( X_train )

# # R²
r2_squared_train_dt = mt.r2_score(y_train, y_pred_train_dt)
print("R²:", r2_squared_train_dt)
# MSE 
mse_train_dt = mt.mean_squared_error(y_train, y_pred_train_dt)
print("MSE", mse_train_dt)
# RMSE
rmse_train_dt = np.sqrt(mse_train_dt)
print("RMSE", rmse_train_dt)
# MAE
mae_train_dt = mt.mean_absolute_error(y_train, y_pred_train_dt)
print("MAE", mae_train_dt)
# MAPE
y_train_flat_dt = y_train
y_pred_train_dt = y_pred_train_dt.ravel()

mask = y_train_flat_dt != 0
mape_train_dt = np.mean(
    np.abs((y_train_flat_dt[mask] - y_pred_train_dt[mask]) / y_train_flat_dt[mask])
)
print("MAPE", mape_train_dt)
```

```python
#Verificando o melhor valor de max_depth para o Decision Tree Regressor
max_depth_values = range(1, 31)
resultados_dt = []
for i in range(1,31):
    model_val_dt = tr.DecisionTreeRegressor(max_depth=i)
    model_val_dt.fit(X_train, y_train)

    y_pred_val = model_val_dt.predict(X_val)
    
    # Cálculo das métricas
    r2 = mt.r2_score(y_val, y_pred_val)
    mse = mt.mean_squared_error(y_val, y_pred_val)
    rmse = np.sqrt(mse)
    mae = mt.mean_absolute_error(y_val, y_pred_val)
    mape = mt.mean_absolute_percentage_error(y_val, y_pred_val)
    
    # Salva os resultados em um dicionário
    resultados_dt.append({
        'max_depth': i,
        'R2': r2,
        'MSE': mse,
        'RMSE': rmse,
        'MAE': mae,
        'MAPE': mape
    })

# Transforma em DataFrame para facilitar a visualização
dt_resultados = pd.DataFrame(resultados_dt)

# Ordena pelo melhor R² ou menor RMSE
dt_resultados = dt_resultados.sort_values(by='R2', ascending=False)
print(dt_resultados)
```

```python
#Previsão com o modelo inicial sobre os dados de validação
y_pred_val_dt = model_train_dt.predict( X_val )
# # R²
r2_squared_val_dt = mt.r2_score(y_val, y_pred_val_dt)
print("R²:", r2_squared_val_dt)
# MSE 
mse_val_dt = mt.mean_squared_error(y_val, y_pred_val_dt)
print("MSE", mse_val_dt)
# RMSE
rmse_val_dt= np.sqrt(mse_val_dt)
print("RMSE", rmse_val_dt)
# MAE
mae_val_dt = mt.mean_absolute_error(y_val, y_pred_val_dt)
print("MAE", mae_val_dt)
# MAPE
y_val_flat_dt = y_val
y_pred_val_dt = y_pred_val_dt.ravel()

mask = y_val_flat_dt != 0
mape_val_dt = np.mean(
    np.abs((y_val_flat_dt[mask] - y_pred_val_dt[mask]) / y_val_flat_dt[mask])
)
print("MAPE", mape_val_dt)
```

```python
#Modelo validado, agora vamos treinar o modelo com os dados de treinamento e validação juntos para depois testar com os dados de teste
#  Model Training
model_last_dt = tr.DecisionTreeRegressor(max_depth=12)
model_last_dt.fit( pd.concat( (X_train, X_val) ), np.concatenate((y_train, y_val)) )
```

```python
#LOOP para investigar os melhores parâmetros do Random Forest Regressor
# Definição dos parâmetros para investigação
n_estimators_list = [50, 100, 200, 500]
max_depth_list = [5, 10, 20, 30]

resultados = []

for n in n_estimators_list:
    for d in max_depth_list:
        # Inicializa e treina o modelo
        model = RandomForestRegressor(n_estimators=n, max_depth=d, random_state=42, n_jobs=-1)
        model.fit(X_train, y_train)
        
        # Predições
        y_pred = model.predict(X_val)
        
        # Cálculo das métricas
        r2 = mt.r2_score(y_val, y_pred)
        mse = mt.mean_squared_error(y_val, y_pred)
        rmse = np.sqrt(mse)
        mae = mt.mean_absolute_error(y_val, y_pred)
        mape = mt.mean_absolute_percentage_error(y_val, y_pred)
        
        # Salva os resultados em um dicionário
        resultados.append({
            'n_estimators': n,
            'max_depth': d,
            'R2': r2,
            'MSE': mse,
            'RMSE': rmse,
            'MAE': mae,
            'MAPE': mape
        })

# Transforma em DataFrame para facilitar a visualização
df_resultados = pd.DataFrame(resultados)

# Ordena pelo melhor R² ou menor RMSE
df_resultados = df_resultados.sort_values(by='R2', ascending=False)
print(df_resultados)
```

```python
#Previsão sobre os dados de teste
y_pred_net_pr_teste = model_last_net_pr.predict( poly.transform(X_test_scaled) )
# # R²
r2_squared_test_net_pr = mt.r2_score(y_test, y_pred_net_pr_teste)
print("R²:", r2_squared_test_net_pr)
# MSE 
mse_test_net_pr = mt.mean_squared_error(y_test, y_pred_net_pr_teste)
print("MSE", mse_test_net_pr)
# RMSE
rmse_test_net_pr = np.sqrt(mse_test_net_pr)
print("RMSE", rmse_test_net_pr)
# MAE
mae_test_net_pr = mt.mean_absolute_error(y_test, y_pred_net_pr_teste)
print("MAE", mae_test_net_pr)
# MAPE
y_test_flat = y_test
y_pred_net_pr_teste_flat = y_pred_net_pr_teste.ravel()

mask = y_test_flat != 0
mape_test_net_pr = np.mean(
    np.abs((y_test_flat[mask] - y_pred_net_pr_teste_flat[mask]) / y_test_flat[mask])
)
print("MAPE", mape_test_net_pr)
```
